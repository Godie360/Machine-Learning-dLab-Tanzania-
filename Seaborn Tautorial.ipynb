```python
## Setup

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
## loading Data
dataset = sb.load_dataset('car_crashes')
dataset.head()
```

```python
 ## Distribution PLot
sb.distplot(dataset['not_distracted'], kde=False, bins=25)   
```

```python
## JOin Plot

sb.jointplot(x= 'speeding', y='alcohol', data=dataset, kind='kde'
```

```python
# KDE Plot

sb.kdeplot(dataset['alcohol'])
```

```python
# Pair plot

# sb.pairplot(dataset)

tips_df= sb.load_dataset('tips')
sb.pairplot(tips_df, hue='sex', palette='Blues')
```

```python
# Rug plot

sb.rugplot(tips_df['tip'])
```

```python
# Styling

sb.set_style('ticks')
plt.figure (figsize=(8,4))

sb.set_context('paper', font_scale=1.4)

sb.jointplot(x='speeding', y='alcohol', data=dataset, kind='reg')
```

## Categorical Data Plotting

```python
# Bar Plot

sb.barplot(x='sex',y='total_bill', data=tips_df, estimator=np.cov)
```

```python
# Count Plot
sb.countplot(x='sex', data=tips_df)
```

```python
## Box Plot

sb.boxplot(x='day', y='total_bill', data= tips_df, hue='sex' )
plt.legend(loc=0)
```

```python
## Violin Plot

sb.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

```python
## Strip Plot

plt.figure(figsize=(8,5))
sb.stripplot(x='day',y='total_bill', data=tips_df, jitter=True, hue='sex',dodge=True)
```

```python
## Swarm Plotviolin
sb.swarmplot(x= 'day', y='total_bill', data=tips_df,color='red', hue='sex', palette='afmhot')
plt.legend(loc=0)
sb.set_style('dark')
```
