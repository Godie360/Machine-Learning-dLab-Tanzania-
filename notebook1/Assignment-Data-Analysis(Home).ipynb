## Checking for Duplicates and removing Duplicates

```python

# Check for duplicates in each column
for column in tourism_data.columns:
    duplicates = tourism_data[column].duplicated().any()
    if duplicates:
        print(f"Column '{column}' --> DUPLICATED.")
    else:
        print(f"Column '{column}' --> NOT DUPLICATED.")

print("")        

check_dup= tourism_data.duplicated().sum()
print ("The List below show that all are duplicated except the UniqueID and this proof that our data duplication is =", check_dup)
```

```python
missing_values_per_column = tourism_data.isna().sum()

total_missing_values = missing_values_per_column.sum()

print("Missing values count in each column:")
print(missing_values_per_column)

print("\nTotal missing values across all columns:", total_missing_values)
```

## Checking the most frequency value that occured in  column  and replacing it into  with missing value

```python
column_to_check = 'most_impressing'
column_to_check2 = 'travel_with'
column_to_check3 = 'total_male'
column_to_check4 = 'total_female'

most_frequent_word = tourism_data[column_to_check].mode()[0]
print(f" The most appeared word in '{column_to_check}' is {most_frequent_word}")
tourism_data[column_to_check].fillna(most_frequent_word, inplace=True)

print('')

most_frequent_word2 = tourism_data[column_to_check2].mode()[0]
print(f" The most appeared word in '{column_to_check2}' is {most_frequent_word2}")
tourism_data[column_to_check2].fillna(column_to_check2, inplace=True)

print('')

most_frequent_word3 = tourism_data[column_to_check3].mode()[0]
print(f" The most appeared word in '{column_to_check3}' is {most_frequent_word3}")
tourism_data[column_to_check3].fillna(most_frequent_word, inplace=True)

print('')

most_frequent_word4 = tourism_data[column_to_check4].mode()[0]
print(f" The most appeared word in '{column_to_check4}' is {most_frequent_word4}")
tourism_data[column_to_check4].fillna(most_frequent_word, inplace=True)
```

```python
missing_values_per_column = tourism_data.isna().sum()

total_missing_values = missing_values_per_column.sum()

print("Missing values count in each column:")
print(missing_values_per_column)

print("\nTotal missing values across all columns:", total_missing_values)
```

## Performing Data Transformation

```python
clean_tourist_data = pd.read_csv('data/CleanTourism.csv')
clean_tourist_data
```

```python
clean_tourist_data.shape
```

```python
aggregated_data = clean_tourist_data.groupby('country').agg({'total_cost':'sum',})
aggregated_data = aggregated_data.sort_values(by='total_cost', ascending=False)
aggregated_data
## The Output Display in series the first up to the last country how have used high cost in Tourism
```

```python
aggregated_data2 = clean_tourist_data.groupby('country').agg({'night_mainland':'sum',})
aggregated_data2 = aggregated_data2.sort_values(by='night_mainland', ascending=False)
total_sum_nights_Tanzania = aggregated_data2['night_mainland'].sum()
print(f"The total Tourist Visit in Mainland are {total_sum_nights_Tanzania}")
aggregated_data2
```

```python
aggregated_data3 = clean_tourist_data.groupby('country').agg({'night_zanzibar':'sum',})
aggregated_data3 = aggregated_data3.sort_values(by='night_zanzibar', ascending=False)
total_sum_nights_zanzibar = aggregated_data3['night_zanzibar'].sum()
print(f"The total Tourist Visit in Zanzibar are {total_sum_nights_zanzibar}")
aggregated_data3
```

```python
max_cost = clean_tourist_data['total_cost'].max()
min_cost = clean_tourist_data['total_cost'].min()

print(f"The Maximum cost is {max_cost} ")
print(f"The Minimum cost is {min_cost} ")
```

```python
clean_tourist_data['level of total cost'] = pd.cut(clean_tourist_data['total_cost'], bins=2, labels=[ 'medium', 'high']) 

columns = ['total_cost','level of total cost']
clean_tourist_data[columns]
```
